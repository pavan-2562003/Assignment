```python
import requests
from bs4 import BeautifulSoup
import csv

def get_product_details(page_url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    response = requests.get(page_url, headers=headers)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        products = soup.find_all('div', {'data-component-type': 's-search-result'})

        product_details = []

        for product in products:
            product_url = 'https://www.amazon.in' + product.find('a', class_='a-link-normal s-no-outline')['href']
            product_name = product.find('span', class_='a-size-medium a-color-base a-text-normal').text.strip()
            product_price = product.find('span', class_='a-price-whole')
            product_price = product_price.text if product_price else 'N/A'
            product_rating = product.find('span', class_='a-icon-alt')
            product_rating = product_rating.text if product_rating else 'N/A'
            product_reviews = product.find('span', {'class': 'a-size-base', 'dir': 'auto'})
            product_reviews = product_reviews.text if product_reviews else 'N/A'

            product_details.append([product_url, product_name, product_price, product_rating, product_reviews])

        return product_details

    else:
        print('Error:', response.status_code)
        return []

def scrape_amazon_products(pages=20):
    base_url = 'https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_{}'

    all_product_details = []

    for page_number in range(1, pages + 1):
        page_url = base_url.format(page_number)
        product_details = get_product_details(page_url)
        all_product_details.extend(product_details)

    return all_product_details

# Save the scraped data to a CSV file
def save_to_csv(data, filename):
    with open(filename, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['Product URL', 'Product Name', 'Product Price', 'Rating', 'Number of Reviews'])
        writer.writerows(data)

if __name__ == '__main__':
    num_pages_to_scrape = 20
    products_data = scrape_amazon_products(pages=num_pages_to_scrape)
    save_to_csv(products_data, 'amazon_products_part1.csv')
```

```python
import requests
from bs4 import BeautifulSoup
import csv

def get_product_details(product_url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    response = requests.get(product_url, headers=headers)
    
    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Initialize variables with 'N/A' values in case the elements are not found
        product_description = 'N/A'
        asin = 'N/A'
        manufacturer = 'N/A'
        
        # Extract the additional information if available
        description_elem = soup.find('div', {'id': 'productDescription'})
        if description_elem:
            product_description = description_elem.text.strip()
        
        asin_elem = soup.find('th', text='ASIN')
        if asin_elem:
            asin = asin_elem.find_next('td').text.strip()
        
        manufacturer_elem = soup.find('th', text='Manufacturer')
        if manufacturer_elem:
            manufacturer = manufacturer_elem.find_next('td').text.strip()

        return [product_url, product_description, asin, manufacturer]

    else:
        print('Error:', response.status_code)
        return []

def fetch_additional_info(product_urls):
    all_product_details = []

    for product_url in product_urls:
        product_details = get_product_details(product_url)
        if product_details:
            all_product_details.append(product_details)

    return all_product_details

if __name__ == '__main__':
    # Sample product URLs (replace with actual product URLs)
    product_urls = ['https://www.amazon.in/dp/B07W3MRXWM', 'https://www.amazon.in/dp/B07VJYZ1ZT']
    
    additional_info_data = fetch_additional_info(product_urls)
    
    # Save the additional information to a CSV file
    with open('amazon_products_part2.csv', mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['Product URL', 'Description', 'ASIN', 'Manufacturer'])
        writer.writerows(additional_info_data)
```
